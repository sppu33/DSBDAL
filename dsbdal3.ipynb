Descriptive Statistics - Measures of Central Tendency and variability

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv("nba.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

# filling null places with measures of central tendency


```python
mode= df['College'].mode()[0]
df['College'].fillna(mode,inplace=True)
mode
```

```python
df['Salary'].fillna(df['Salary'].mean(), inplace=True)
```

```python
df['College'].value_counts()
```

```python
collegeGroups=df.groupby(df['College'])
```

```python
collegeGroups.describe()
```

```python
meanSalaryByCollege=collegeGroups['Salary'].mean()
```

```python
meanSalaryByCollege
```

converting Salary Values to List

```python
incomeListByCollege=collegeGroups['Salary'].apply(list)
incomeListByCollege
```

# to categorize (or bin) the "Age" column into age groups.

```python
bins=[20,25,30,35,40]
labels=['20-24','25-29','30-34','35-49']
ageGroups=pd.cut(df['Age'],bins=bins, labels=labels, right=False)
```

```python
ageGroups.value_counts()
```

```python
df.head()
```

```python
df['Age Group']=ageGroups
```

```python
df.head()
```

```python
SalaryListByAgeGroup=df.groupby('Age Group')['Salary'].apply(list)
SalaryListByAgeGroup
```
